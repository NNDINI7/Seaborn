```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Facet_grid is known as multiple plots,where we can plot multiple plots in a single grid.

```python
df = sns.load_dataset("tips")
df
```

```python
fg=sns.FacetGrid(df)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
#to create multigraph make use of col parameter ie. name of attributes in df
fg=sns.FacetGrid(df,col="sex")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(df,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()#add_legend is used to add legend 
plt.show()
```

```python
fg=sns.FacetGrid(df,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()#add_legend is used to add legend 
plt.show()
```

```python
fg=sns.FacetGrid(df,col="sex",hue="day")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
