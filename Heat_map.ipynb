```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = np.linspace(1,10,20).reshape(4,5)
df
```

```python
sns.heatmap(df)
```

```python
# using multiple parameters 
sns.heatmap(df, vmin=0, vmax=12 )
#max and min values
```

```python
sns.heatmap(df,cmap="viridis")
#cmap changes color of the map
```

```python
sns.heatmap(df, annot = True)
```

```python
arr = np.linspace(1,10,10).reshape(2,5)
#values from 1 to 10 , with 10 numbers 
arr
```

```python
sns.heatmap(arr, annot=True)
```

```python
#creating array 
ar=np.array([["a0","a1","a2","a3","a4"],
            ["b0","b1","b2","b3","b4"]])
ar
```

```python
u={"fontsize":10,"color":"yellow"}
sns.heatmap(arr, annot=ar ,fmt="s",annot_kws=u)
#fmt is format of the values i.e. string = "s".
#annot_kws is used to change the size and color of the values.
```

```python

sns.heatmap(arr, annot=ar ,fmt="s",linewidth=10,linecolor ="black")
```

```python

sns.heatmap(arr, annot=ar ,fmt="s",cbar = False)
#cbar is used to remove color bar
plt.show()
```

```python

l=sns.heatmap(arr, annot=ar ,fmt="s",linewidth=10)
l.set(xlabel="python",ylabel="java") #assigns label to axis
plt.show()
```
