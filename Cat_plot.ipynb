```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df= sns.load_dataset("tips")
df
```

```python
#plotting factor plot 
y=df.head(20)
sns.catplot(x="size",y="tip",data=y)
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=y,hue="sex",palette="plasma")
plt.show()
```

```python
#using height parameter
sns.catplot(x="size",y="tip",data=y,hue="sex",palette="plasma",height=10)
plt.show()
```

```python
#using kind parameter
sns.catplot(x="day",y="size",data=y,hue="sex",palette="plasma", kind="bar")
plt.show()
```

```python
#using kind parameter
sns.catplot(x="day",y="size",data=y,
            hue="sex",palette="plasma", kind="boxen")
plt.show()
```

```python
#using kind parameter
sns.catplot(x="day",y="size",data=y,
            hue="sex",palette="plasma", kind="box")
plt.show()
```

```python
#using kind parameter
sns.catplot(x="day",y="size",data=y,
            hue="sex",palette="plasma", kind="point")
plt.show()
```

```python
#using kind parameter
sns.catplot(x="day",y="size",data=y,
            hue="sex",palette="plasma", kind="swarm")
plt.show()
```
