```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset("tips")
df
```

```python
# plotting box plot
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python

sns.boxplot(x="day",y="total_bill",data=df,hue="time",palette="viridis")
plt.show()
```

```python

sns.boxplot(x="day",y="total_bill",data=df,hue="sex",palette="viridis")
plt.show()
```

```python
#using order parameter
sns.boxplot(x="day",y="total_bill",data=df,hue="time",palette="viridis",order=["Fri","Sun","Thur","Sat"])
plt.show()
```

```python
#showing mean parameter
sns.boxplot(x="day",y="total_bill",data=df,hue="time",palette="viridis",showmeans=True)
plt.show()
```

```python
#change mean property

sns.boxplot(x="day",y="total_bill",data=df,hue="time",
            palette="plasma",showmeans=True,meanprops={"marker":"+","markeredgecolor":"black"}
        )
plt.show()
```

```python
#plotting single data variable
sns.boxplot(x=df["total_bill"])
```
