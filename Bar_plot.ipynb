```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

```

```python
df = sns.load_dataset("titanic")
df
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df)
plt.title('Survival Rate by Class')
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df,hue="who")
plt.title('Survival Rate by Class using hue')
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df, order= ['Third', 'Second', 'First'])
plt.title('Survival Rate by Class using order')
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df ,hue="who",hue_order=["child","woman","man"] )
plt.title('Survival Rate by Class')
plt.show()
#hue_order is used to order the hue categories in the plot
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df, ci=1)
# ci=100 will show the confidence interval as 100% which means no error bars
# This is not a common use case, but it demonstrates how to set the confidence interval.

plt.title('Survival Rate by Class')
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df, n_boot=1000, seed=42, hue="who")
# n_boot=1000 will perform 1000 bootstrap resamples to estimate the confidence intervals
# This is useful for more robust statistical analysis.
#seed=42 ensures reproducibility of the results.
plt.title('Survival Rate by Class')
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df , hue="who",orient='h')
plt.title('Survival Rate by Class with Horizontal Orientation')
plt.show()

#orient='h' is used to create a horizontal bar plot
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df , color="purple")
plt.title('Survival Rate by Class')
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df,hue="who",palette="viridis")
plt.title('Survival Rate by Class')
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df,hue="who",palette="viridis", saturation= 20)
plt.title('Survival Rate by Class')
# saturation=20 will increase the color intensity of the bars.
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df,hue="who",palette="viridis", errcolor='red', errwidth=15)
plt.title('Survival Rate by Class')
# errcolor='red' will change the color of the error bars to red.
# errwidth=2 will increase the width of the error bars.
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df,hue="who",palette="viridis", errcolor='red', capsize=1)
plt.title('Survival Rate by Class')
# capsize=1 will add caps to the error bars. gives endpoint of the error bars.
plt.show()
```

```python
# Create a bar plot
sns.barplot(x='class', y='survived', data=df,hue="who",palette="viridis",dodge=False, alpha=0.8)
plt.title('Survival Rate by Class')
plt.show()
# dodge=False will group the bars together without separating them by hue.
# alpha=0.8 will set the transparency of the bars to 80%, making them slightly transparent.it blurs the bar slightly.
```
