```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Load the diamonds dataset
df = sns.load_dataset("diamonds")
df
```

```python
sns.displot(df["carat"])
plt.title("Distribution of Diamond Carat")
plt.show()
```

```python
sns.displot(df["price"],bins=20)
#bins is used to specify the number of bins in the histogram
plt.title("Distribution of Diamond Price")
```

```python
sns.displot(df["price"],bins=20,kde = True,rug=True)
# kde=True adds a kernel density estimate to the histogram
# rug=True adds a rug plot to the histogram
plt.title("Distribution of Diamond Price with KDE")
plt.show()
```

```python
sns.displot(df["cut"],bins=20,color="purple")
plt.show()
```

```python
x = df.head(20)  # Display the first 20 rows of the dataset
```

```python
#i want to display a histogram with legends on different values of cut 
#make use of the cut column in the dataset
sns.displot(x, x="cut", hue="cut", multiple="stack", palette="Set1")
plt.title("Distribution of Diamond Cut with Legends")   
plt.show()

```
