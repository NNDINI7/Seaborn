```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1,2,3,4,5,6,7]
var1 = [2,4,3,1,4,5,6]
df = pd.DataFrame({"var": var, "var1": var1})   

sns.lineplot(x="var", y="var1" , data = df)
plt.title("Line Plot Example")
plt.show()


```

```python
data_store= sns.load_dataset("penguins")
data_store
```

take dataset 


```python

sns.lineplot(x="bill_length_mm", y="bill_depth_mm" , data = data_store)
plt.title("Penguins Bill Length vs Depth")
plt.show()


```

```python

sns.lineplot(x="bill_length_mm", y="bill_depth_mm" , data = data_store,hue="species")
plt.title("Penguins Bill Length vs Depth")
plt.show()

#hue is used to show different categories in the data with respect to species


```

```python

sns.lineplot(x="bill_length_mm", y="bill_depth_mm" , data = data_store,hue="sex", size=50)
plt.title("Penguins Bill Length vs Depth")
plt.show()

#size is used to show the size of the points in the plot, here it is set to 50
#hue is used to show different categories in the data with respect to sex
```

```python

sns.lineplot(x="bill_length_mm", y="bill_depth_mm" , data = data_store,hue="sex",style="sex" )
plt.title("Penguins Bill Length vs Depth")
plt.show()

#style is used to show different categories in the data with respect to sex
```

```python

sns.lineplot(x="bill_length_mm", y="bill_depth_mm" , data = data_store,hue="sex", palette='viridis')
plt.title("Penguins Bill Length vs Depth")
plt.show()

#palette is used to change the color scheme of the plot
```

```python

data_store = sns.load_dataset("penguins").head(20)
data_store
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm" , data = data_store,hue= "sex",style="sex",markers=["o",">"])
#markers=["o",">"] is used to specify different markers for different categories
plt.title("Penguins Bill Length vs Depth")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm" , data = data_store,hue= "sex",style="sex",markers=["o",">"], legend = False)

#legend=False is used to hide the legend in the plot

plt.title("Penguins Bill Length vs Depth")
plt.show()
```
